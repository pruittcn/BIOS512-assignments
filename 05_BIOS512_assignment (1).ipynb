Each row in the table represents a single turbine.

```r
wind <- read.csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')
summary(wind)
```

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
p = ggplot(wind, aes(y=forcats::fct_infreq(province_territory))) +
    geom_bar()
p = p + labs(y= "Province/Territory", x= "Number of Wind Turbines")
p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind2 = wind |>
mutate(manu_lumped=fct_infreq(manufacturer) |>
                   fct_lump_lowfreq() |>
                   fct_relevel('Other'))


p = ggplot(wind2, aes(y=manu_lumped)) + geom_bar()
p = p + labs(y= "Manufacturer", x= "Number of Wind Turbines")
p = p + theme_classic()
p
```
